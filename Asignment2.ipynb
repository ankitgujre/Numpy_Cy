```python
# Q.5 Write a code to edit the value 40 by 400 in given 1D array
# Input: A=[20,30,40,50,60]
A = np.array([20,30,40,50,60])
print(A)
A[2] = 400
print(A)
```

```python
A[A == 40] = 400
print(A)
```

```python
A = np.where(A == 40,400, A)
print(A)
```

```python
A = np.where(A == 40, 400, A)
print(A)
```

```python
A = np.array([20, 30, 40, 50, 60])  # Fresh array
print(A)  # [20 30 40 50 60]

A = np.where(A == 40, 400, A)
print(A)
```

```python
# Q.7 Write a code to delete 50 in given 1D array.
# Input: A=[20,30,40,50,60]
A = np.array([20,30,40,50])
print(A)
A = np.delete(A, np.where(A==30))
print(A)
```

```python
# Q.9 Write a code to find odd number in given 1D array.
# Input: A=[20,3,7,80,89]
A = np.array([20,3,7,80,89])
print(A)
np.where(A % 2 != 0)
```

```python
import numpy as np
import pandas as pd
```

```python
# wap for mean

data = np.array([100,200,350,200,400,500,200,600])

total_sum = 0
count = 0

for i in data:
    total_sum += i
    count += 1

mean = total_sum / count

# print(f"Mean: {mean}")
print("Mean = ",mean.round(2))
```

```python
# wap for median
sorted_data = np.sort(data)
print(sorted_data)
n = len(sorted_data)
if n%2 == 1:
    median = sorted_data[n//2]
else:
    median = (sorted_data[n//2-1] + sorted_data[n//2])/2

print("median = ", median)
```

```python
# wap for mode
frequency_dict = {}

# 1. Count frequencies using a loop
for i in data:
    if i in frequency_dict:
        frequency_dict[i] += 1
    else:
        frequency_dict[i] = 1

# 2. Find the item with the maximum frequency
max_count = max(frequency_dict.values())
mode = []

for key, value in frequency_dict.items():
    if value == max_count:
        mode.append(key)

print(f"Frequency Map: {frequency_dict}")
print(f"Mode: {mode} (appears {max_count} times)")
```

```python
# imputation of data
arr = np.array([10,20,np.nan,40,50,np.nan,70])
print(arr)
```

```python
# calculate mean ignorinng nan
mean_value = np.nanmean(arr)
# print(mean_value)
# replace nan with mean
arr_mean_imputed = np.where(np.isnan(arr) , mean_value, arr)
print("Mean=", mean_value)
print(arr_mean_imputed)
```

```python
# same for median
import numpy as np

# imputation of data
arr = np.array([10, 20, np.nan, 40, 50, np.nan, 70])
print("Original Array:", arr)

# 1. calculate median ignoring nan
median_value = np.nanmedian(arr)

# 2. replace nan with median
arr_median_imputed = np.where(np.isnan(arr), median_value, arr)

print("Median =", median_value)
print("Imputed Array:", arr_median_imputed)
```

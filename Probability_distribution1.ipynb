```python
import numpy as np
```

```python
# Probability Distribution

# Binomial Distribution using numpy

np.random.seed(5)
heads = np.random.binomial(
    n = 10, 
    p = 0.5,
    size = 15
)

print("Heads in each experiment:", heads)
print("\n Average heads = ", np.mean(heads))
```

```python
# Poission Distribution

np.random.seed(20)
calls = np.random.poisson(
    lam = 4,
    size = 20
)

print("Call per minutes:")
print(calls)
print("\n Average call = ", np.mean(calls))
print("\n Maximum calls = ", np.max(calls))
print("\n Minimum calls = ", np.min(calls))
```

```python
# uniform distribution

np.random.seed(40)
numbers = np.random.uniform(
    low = 10,
    high = 20,
    size = 15
)

print(np.round(numbers, 2))
print("\n Average numbers = ", np.round(np.mean(numbers)))
print("\n Maximum numbers = ", np.max(numbers))
print("\n Minimum numbers = ", np.min(numbers))
```

```python
# Normal Distribution

import numpy as np
np.random.seed(10)
marks = np.random.normal(
    loc = 70,
    scale = 8,
    size = 20
)
print("Generated Marks:")
print(np.round(marks, 2))
print("\n Average marks = ", np.round(np.mean(marks)))
print("\n Maximum marks = ", np.max(marks))
print("\n Minimum marks = ", np.min(marks))
```

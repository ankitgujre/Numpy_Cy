```python
# 1. Create a 2D array with 1 on border and 0 on inside.

import numpy as np
```

```python
arr = np.ones((10,10), dtype = int)
print(arr)
```

```python
arr[1:-1, 1:-1] = 0 
print(arr)
```

```python
# Find indices of Non-zero elements from A[2,3,0,1,0,9].
A = np.array([2,3,0,1,0,9])
arr = np.nonzero(A)
print(arr)
print(A[arr])
```

```python
# Create a zero one D vector of size 10 but 5 th indexing value should be replaced by 9.
Arr = np.zeros(10, dtype = int)
Arr[5] = 9
print(Arr)
```

```python
# Create a 3×3 NumPy array of all True.
arr = np.ones((3, 3), dtype=bool)
print(arr)
```

```python
# Create a 4×4 array and extract its first row and last column.
arr = np.arange(1,17)
```

```python
print(arr)
```

```python
arr = arr.reshape(4,4)
print(arr)
```

```python
print(arr[0, :])
print(arr[:, -1])
```

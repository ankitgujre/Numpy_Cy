```python
import numpy as np
```

```python
# Q.1 Write a NumPy program to create an array of integers from 30 to 70.

arr1 = np.arange(30,71)
print(arr1)
```

```python
# Q.2 Write a NumPy program to create an array of all even integers from
# 40 to 139.

arr2 = np.arange(40,140,2)
arr2
```

```python
# Q.3 Write a NumPy program to create an array of 10 zeros, 10 ones, and
# 10 fives.

zeros = np.zeros(10)
ones = np.ones(10)
fives = np.full(10, 5)

print(zeros)
print(ones)
print(fives)
```

```python
# Q.4 Write a NumPy program to create a 3x3 identity matrix.

identity = np.eye(3, dtype = int)
print(identity)
```

```python
# Write a NumPy program to create a given matrix.

arr3 = np.arange(1,21)
arr3
arr3.reshape(5,4)
```

```python
# Q.6 Write a NumPy program to print above matrix in reverse order.(20 to
# 1)

arr4 = np.arange(20,0,-1)
print(arr4)
```

```python
# Write a program to pick any 3 random values from 100 to 300.

rd = np.random.randint(100,300,3)
print(rd)
```

```python
# Q.8 Write a program to pick any random three names from given List

A=['SACHIN','VIRAT','DHONI','ROHIT','DRAVID','GANGOLY']

a = np.array(A)
rd = np.random.choice(a,3)
print(rd)



```

```python
import numpy as np
# outliers handling by IQR
```

```python
data = np.array([13,16,19,22,23,27,30,33,37,45])
print(data)
```

```python
mean = np.mean(data)
print(mean)
```

```python
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_Bound = Q1  - 1.5*IQR
Upper_Bound = Q3+1.5*IQR

Outliers = data[(data < lower_Bound) | (data > Upper_Bound)]
print("Q1:", Q1)
print("Q3:",Q3)
print("IQR :", IQR)
print("Outliers:", Outliers)
```
